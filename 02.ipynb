```python
import pandas as pd
import re
import nltk
```

```python
# usamos o set_option do pandas para customizar as configurações originais
pd.set_option('display.max_colwidth', 100)
rowData = pd.read_table('SMSSpamCollection.tsv', names=['H or S','Text'])
rowData.head(20)
```

```python
rowData['clean'] = rowData.Text.astype(str).str.replace(',' , "") 
rowData['clean'] = rowData.clean.astype(str).str.replace('.' , "") 
rowData['clean'] = rowData.clean.astype(str).str.replace('?' , "") 
rowData['clean'] = rowData.clean.astype(str).str.replace(')' , "")
rowData['clean'] = rowData.clean.astype(str).str.replace('/' , "")
rowData['clean'] = rowData.clean.astype(str).str.replace(':' , "")
rowData['clean'] = rowData.clean.astype(str).str.replace('!' , "")
rowData['clean'] = rowData.clean.astype(str).str.replace("'" , "")
rowData['clean'] = rowData.clean.astype(str).str.replace("¨" , "")     
rowData['clean'] = rowData.clean.astype(str).str.replace("‘" , "")    
```

```python
rowData.head(20)
```

```python
# another way to clean pontuation
import string
string.punctuation
```

```python
# criando uma funcnao para remover as pontuações
# o join é usado pra juntar os carecteres, antes da funçao nos especificamos como sera juntado
def remove_punct(text):
    text_nopunct = ''.join([char for char in text if char not in string.punctuation])
    return text_nopunct
```

```python
# removendo todos os pontos da liste de outra forma
rowData['clean2'] = rowData.Text.apply(lambda x: remove_punct(x))
rowData.head(20)
```

```python
# criando uma nova funçnao para separar todas as palavras usando W que separa tudo que não é palavra
def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens

rowData['tokenize'] = rowData.clean2.apply(lambda x: tokenize(x.lower()))
rowData.head(10)
```

```python
# usando a funcao do nltk para removendo palavras irrelevantes ou Remove stopwords
stopword = nltk.corpus.stopwords.words('english')

# criando uma funçnao para verificar se existe palavras irrelevantes na lista tokenize
def remove_stopword(tokenize):
    text = [word for word in tokenize if word not in stopword]
    return text
```

```python
# usando lambda para verificar se existe alguma palavra irrelavante
rowData['noStopword'] = rowData.tokenize.apply(lambda x: remove_stopword(x))
rowData.head(20)
```
