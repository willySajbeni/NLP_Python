#### TF - IDF 
##### 'frase que vou usar de EXEMPLO'
##### Formula é Q_Vz_EXMPLO_Aparece / N_Words_in_text
##### depois fazemos os log(numeor de frase / numero de vezes que EXEMPLO apareceu)
##### nesse exemplo vou supor que tivemos 20 frases e em apenas uma aparece EXEMPLO
### $$  = 1/6  $$
### $$  = 1/6*log(20/1)  $$

```python
import pandas as pd 
import re
import string
import nltk
pd.set_option('display.max_colwidth', 100)
# aqui impotamos o TfildfVectorize para usarmos na vetorizacao
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# criando o data
data = pd.read_table('SMSSpamCollection.tsv', names=['S or H', 'Text'])
```

```python
data.head()
```

```python
# criando filtros de palavras irrelavantes e criando objeto ps para dar um stem nas palvras
stopword = nltk.corpus.stopwords.words('english')
ps = nltk.PorterStemmer()
```

```python
# criando funcao para tirar pontucao, tokenize, filtrar usando stopword e stimming
def clean_text(text):
    text = ''.join([word for word in text if word not in string.punctuation])
    tokenize = re.split('\W+', text)
    text = ' '.join([ps.stem(word) for word in tokenize if word not in stopword])
    return text

```

```python
# criando a coluna limpa
data['clean_text'] = data.Text.apply(lambda x : clean_text(x.lower()))
data.head(5)
```

```python
# crete a small set
data_sample = data[0:20]
data_sample.head(20)
```

```python
# vamos vetorizar o data simplificado criado a cima 
tfid_vect = TfidfVectorizer(analyzer=clean_text)
X_tdif = tfid_vect.fit_transform(data_sample.clean_text)
print(X_tdif.shape)
print(tfid_vect.get_feature_names())
```

```python
# aqui nessa vetorizacao quanto mais perto do 1 mais vezes aparece 
X_df_vector = pd.DataFrame(X_tdif.toarray(), columns=tfid_vect.get_feature_names())
X_df_vector
```
