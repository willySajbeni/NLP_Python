### Using Stemming - clening - aqui reduzimos as palavras a sua verçnao mais raiz 

```python
import nltk
import pandas as pd
import re
import string
pd.set_option('display.max.colwidth', 100)
```

```python
# criando um objeto usando o Stemming do NLTK
ps = nltk.PorterStemmer()
```

```python
# criando Stopword ou palavras irrelavantes para tira-las da lista
stopword = nltk.corpus.stopwords.words('english')
```

```python
dir(ps)
```

```python
print(ps.stem('grows'))
print(ps.stem('growing'))
print(ps.stem('grow'))
```

```python
# criando o data frame 
data = pd.read_table('SMSSpamCollection.tsv', names=['S or H','Text'])
data2 = data.copy()
data.head()
```

```python
# criando uma função para limpar a coluna do dataframe escolhido
def clean_text(text):
    text = ''.join([word for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [word for word in tokens if word not in stopword]
    return text
```

```python
# criando um nova coluna para salvar a lista de palavras peneradas que usaremos
data['noStop'] = data.Text.apply(lambda x: clean_text(x.lower()))
data.head()
```

```python
# criando uma função para usar o metodo .stem
def stemming(tokenized_text):
    text = [ps.stem(word) for word in tokenized_text ]
    return text
```

```python
# usando a função criada a cima + lambda para armazenar e criar a nova coluna
data['stem_text'] = data.noStop.apply(lambda x: stemming(x))
data.head(10)
```

### Agora vamos usar o metodo Lemmatizer

```python
# esse metodo é mais lento, porem, é mais acurado ele procura todos os sinonimos das palavras
# e tenta agrupalos para ter um resultado mais perfeito - WordNet
wn = nltk.WordNetLemmatizer()
```

```python
dir(wn)
```

```python
# criando a funçnao de limpeza novamente por pratica
def clena_text2(text):
    text = ''.join([word for word in text if word not in string.punctuation])
    tokenized = re.split('\W+',text)
    text = [word for word in tokenized if word not in stopword]
    return text
```

```python
data2['noStop'] = data2.Text.apply(lambda x: clena_text2(x.lower()))
data2.head(10)
```

```python
# Criando a funcao para usar o lemmatize
def lemmatize_text(noStop):
    text = [wn.lemmatize(word) for word in noStop]
    return text
```

```python
# criando a nova coluna usando a funcao lemmatize
data2['text_lemmatize'] = data.noStop.apply(lambda x: lemmatize_text(x))
data2.head(10)
```
